```python
import uc2rest

mPort ="/dev/cu.usbserial-A50285BI"
ESP32 = uc2rest.UC2Client(serialport=mPort, DEBUG=True)
# setting debug output of the serial to true - all message will be printed
ESP32.serial.DEBUG=True
```

```python
# move and measure
range = 4000
print("Current position: "+ str(ESP32.motor.get_position(axis=1)))
ESP32.motor.move_z(steps=-range/2, speed=10000, is_blocking=True, is_absolute=False, is_enabled=True)
print("Current position: "+ str(ESP32.motor.get_position(axis=1)))
ESP32.motor.move_z(steps=range, speed=100, is_blocking=True, is_absolute=False, is_enabled=True)
print("Current position: "+ str(ESP32.motor.get_position(axis=1)))
ESP32.motor.move_z(steps=-range/2, speed=10000, is_blocking=True, is_absolute=False, is_enabled=True)
```
